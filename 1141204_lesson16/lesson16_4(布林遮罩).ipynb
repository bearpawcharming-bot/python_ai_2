```python
import pandas as pd

# 建立一個 Series（產品銷售額）
sales = pd.Series([300, 600, 450, 800, 200, 700], 
                  index=['A', 'B', 'C', 'D', 'E', 'F'])

sales
# 建立布林遮罩：銷售額 > 500
mask = sales > 500
print("布林遮罩：")
print(mask)
# 輸出：
# A    False
# B     True
# C    False
# D     True
# E    False
# F     True
# dtype: bool

# 使用遮罩篩選高銷售產品
high_sales = sales[mask]
print("\n銷售額 > 500 的產品：")
print(high_sales)
# 輸出：
# B    600
# D    800
# F    700
# dtype: int64
```
