```python
#lambda 範例-> 暱稱函式
import pandas as pd
s = pd.Series([20, 21, 12], index=['London', 'New York', 'Helsinki'])
s.apply(lambda x: x ** 2)
```

```python
# 使用具名函式
def add_value(x, a, b=1):
    return x + a + b
    
s.apply(add_value, args=(10,), b=5)
```

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2], 'B': [10,20]})
df.apply(sum, axis=0)
df
```

```python
def my_func(row):
    return row['A'] + row['B']
df.apply(my_func, axis=1)
```
