```python
import numpy as np
import pandas as pd

students = np.random.randint(50, high=101, size=(50, 5))
students_df = pd.DataFrame(data=students,
            index=[f"學生{i}" for i in range(1,51)],
            columns=['國文','英文','數學','地理','歷史'])
students_df
```

```python
# 透過欄的選取
students_df['國文']
students_df[['英文','地理']]
```
