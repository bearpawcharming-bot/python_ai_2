```python
import numpy as np
import pandas as pd

students = np.random.randint(50, high=101, size=(50, 5))
students_df = pd.DataFrame(data=students,
            index=[f"學生{i}" for i in range(1,51)],
            columns=['國文','英文','數學','地理','歷史'])
students_df
```

```python
new_students = students_df.copy()
sum_series = np.sum(students_df,axis=1)
mean_series = np.mean(students_df,axis=1)
new_students['總分'] = sum_series
new_students['平均'] = mean_series
new_students
```
