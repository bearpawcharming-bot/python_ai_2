```python
start_date = "2025-01-01"
end_date = "2025-12-31"
tickers = ['2330.TW', '2303.TW', '2454.TW', '2317.TW']
data = {ticker: yf.download(ticker, start=start_date, end=end_date, interval='1d', auto_adjust=True) for ticker in tickers}
data
```

```python
import pandas as pd
data1 = {key:dataframe['Close'][key] for key,dataframe in data.items()}
price_df = pd.DataFrame(data1)
price_df
```

```python
import pandas as pd
data1 = {key:dataframe['Close'][key] for key,dataframe in data.items()}
price_df = pd.DataFrame(data1)
price_df1 = price_df.rename(columns ={'2330.TW':'台積電','2303.TW':'聯電','2454.TW':'聯發科','2317.TW':'鴻海'})
# .rename()更改欄位名稱
price_df1
```

```python
# 相關係數矩陣
price_df1.corr()
```

```python
# 要了解每日的漲跌情況，可以使用 pct_change() 方法計算每日的百分比變化：
daily_returns = price_df1.pct_change()

# 裏面的值,乘以100就是百分比
daily_returns_percentage = daily_returns * 100
daily_returns_percentage
# 把所有是nan的值移除
daily_returns_percentage = daily_returns_percentage.dropna()
# 裏面的數值後要加上百分比符號 %
daily_returns_percentage.style.format("{:.2f}%")
```

```python
# 要了解每日的漲跌情況，可以使用 pct_change() 方法計算每日的百分比變化：
daily_returns = price_df1.pct_change()

# 裏面的值,乘以100就是百分比
daily_returns_percentage = daily_returns * 100
daily_returns_percentage
# 把所有是nan的值移除
daily_returns_percentage = daily_returns_percentage.dropna()
# 裏面的數值後要加上百分比符號 %
#daily_returns_percentage.style.format("{:.2f}%")
# 儲存格內的值如果是負數就變綠色,正數就變紅色
def color_negative_red(value):
    if value < 0:
        color = 'green'
    elif value > 0:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'

# index內的日期格式,只要顯示日期,不顯示時間
daily_returns_percentage.index = daily_returns_percentage.index.date
daily_returns_percentage.style.map(color_negative_red).format("{:.2f}%")
```
