```python
# 數據資料圖表化
# 線的預設是藍色實線

import matplotlib.pyplot as plt #引入matplotlib.pyplot模組，並簡稱為plt

# 1.製作figure
fig = plt.figure() #在網路上查找資料 https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1) #.add_subplot指的是方法

# 3.將圖的數據儲存在axes中
X = [0, 1, 2] #0, 1, 2是x軸的tick位置對應的數值
Y = [3, 4, 5] #3, 4, 5是tick位置對應的數值
ax_1.plot(X, Y) #.plot指的是方法
# 網上資料 https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html
# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt #引入matplotlib.pyplot模組，並簡稱為plt 
# 1.製作figure
fig = plt.figure() 

# 2.做1*3的矩陣圖的第一個axes
ax_1 = fig.add_subplot(1, 3, 1)

# 設定ax_1的標題
ax_1.set_title(label='Ax 1')
# 查資料 https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html 

# 3.做1*3的矩陣圖的第二個axes
ax_2 = fig.add_subplot(1, 3, 2)

# 設定ax_2的標題
ax_1.set_title(label='Ax 2')

# 4.做1*3的矩陣圖的第三個axes
ax_3 = fig.add_subplot(1, 3, 3)

# 設定ax_3的標題
ax_1.set_title(label='Ax 3')

```

```python
import matplotlib.pyplot as plt
# 1.製作figure
fig_1 = plt.figure()
# 2.在1.的figure上製作一個axes
ax_1 = fig_1.add_subplot(1, 1, 1)
# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

fig_2 = plt.figure()
ax_2 = fig_2.add_subplot(1, 1, 1)
X = [2, 3, 6]
Y = [2, 1, 2]
ax_2.plot(X, Y)

# 4.顯示圖
plt.show()
```
