```python
df = pd.read_csv('各鄉鎮市區人口密度.csv') # 讀取 CSV 檔案
# 取出欄位名稱
cols = df.columns.tolist()
cols
```

```python
# 如何更改df的欄位名稱
# 使用已存在的 cols 列表重新命名欄位

df.columns = df.loc[0].tolist()
# 刪除第一列，因為它已經成為新的欄位名稱
df1 = df.drop(index=0).reset_index(drop=True) # 重設索引
# 顯示前10列
df1.head(10)

# 顯示後10列
df1.tail(10) 

# 如何刪除index(370)以後的欄位
# df2 = df1[df1.index <= 370 ]-->ai方式

# 老師教的方式
df1.drop(index=[370,371,372,373,374])
df2 = df1.reindex(index=range(0,370))
df2
```

```python
#ai方式

# 查詢所有df2欄位的資料型態
df2.dtypes
# 如何將年底人口欄位轉成整數型態
# 先移除非數字字元（包含中文省略號等），再轉為數值，最後轉為 pandas 的 nullable integer dtype
df2['年底人口數'] = pd.to_numeric(
    df2['年底人口數'].astype(str).str.replace(r'[^\d]', '', regex=True),
    errors='coerce'
).astype('Int64')
df2
```

```python
# 轉成excel檔案需要 openpyxl 套件

import sys, subprocess
import importlib

# 如何將df2存成新的csv檔案
df2.to_csv('各鄉鎮市區人口密度_清理後.csv', index=False)
# 如何將df2存成新的excel檔案
df2.to_excel('各鄉鎮市區人口密度_清理後.xlsx', index=False)
# 確保 openpyxl 已安裝（若未安裝則透過 pip 安裝）
try:
    import openpyxl  # type: ignore
except ModuleNotFoundError:
    print("openpyxl not found. Installing...")
```

```python
# 老師教的方式
# 應該像上面那個圖，但目前不知道怎麼解決...的問題(羅怡琳)

#df2['年底人口數'].astype(int)
# df2['年底人口數']內容有`… `,無法直接轉成整數,如何置換為0
df2['年底人口數'] = df2['年底人口數'].replace('... ', '0').astype(int)
df2['人口密度'] = df2['人口密度'].replace('... ', '0').astype(int)
df2['土地面積'] = df2['土地面積'].astype(float)
df2.dtypes
```

```python
#df2['土地面積']的內容是float型態,如何只顯示小數點後3位
df2['土地面積'] = df2['土地面積'].round(3)
df2
```
