```python
from tools import all_data_concat
import pandas as pd 

all_data = all_data_concat()
# 欄位`乘車日期`是數字格式,要轉成日期格式
all_data['乘車日期'] = pd.to_datetime(all_data['乘車日期'], format='%Y%m%d')

# 取出日期的年份,建立一個新的欄位`年份`
all_data['年份'] = all_data['乘車日期'].dt.year

# 取出期期的月份,建立一個新的欄位`月份`
all_data['月份'] = all_data['乘車日期'].dt.month
all_data.head()
```

```python
group = all_data.groupby(['車站名稱','年份','月份'])[['進站人數','出站人數']]
group.agg(['mean','sum'])
```

```python
def top(df,n=3,column='進站人數'):
    return df.sort_values(by=column,ascending=False).head(n)
group.apply(top)
```
