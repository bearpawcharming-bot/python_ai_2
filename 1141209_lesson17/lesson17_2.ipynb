```python
import pandas as pd
world_data = pd.read_csv('world.csv')
print(world_data.columns)
print(world_data.info())
```

```python
#reindex 重新排列欄位順序
world_data1 = world_data.reindex(columns=['洲名', '國家', '日期', '面積', '人口密度'])
world_data1
```

```python
# 使用布林遮罩篩選出「國家」欄位為「台灣」的資料

mask_taiwan= world_data1['國家']=='台灣' # 篩選出國家是台灣的列
world_data_taiwan = world_data1[mask_taiwan]  # 取得台灣的資料
world_data_taiwan
```

```python
#dataframe的實體裡有query方法，可以直接用query方法來篩選資料
# df.query(expr, inplace=False, **kwargs)
# expr: 欄位名稱(字串)
# inplace: 是否直接在原本的dataframe做修改，預設為False

world_data_taiwan= world_data1.query("國家=='台灣'")
world_data_taiwan
```

```python
# 篩選出「國家」欄位為「台灣」且「新增確診數」欄位為0的資料
taiwan1 = world_data.query("國家 == '台灣' and 新增確診數 == 0")
taiwan1
```

```python
# 找出台灣「新增確診數」的最高值，並篩選出該筆資料
taiwan_max = world_data_taiwan['新增確診數'].max()
taiwan_max_row = world_data_taiwan.query("新增確診數 == @taiwan_max")
taiwan_max_row
```
