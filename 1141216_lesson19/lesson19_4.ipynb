```python
import pandas as pd

final_dataFrame['乘車日期'] = pd.to_datetime(final_dataFrame['乘車日期'], format='%Y%m%d')
final_dataFrame.head()
```

```python
final_dataFrame.groupby("車站名稱")[["進站人數","出站人數"]].sum()
```

```python
# 如何groupby多欄位,依據年份groupby
final_dataFrame['year'] = final_dataFrame['乘車日期'].dt.year
df1 = final_dataFrame.groupby(['year',"車站名稱"])[["進站人數","出站人數"]].sum()
df1
```
