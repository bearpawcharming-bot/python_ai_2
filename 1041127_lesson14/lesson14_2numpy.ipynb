```python
## numpy
my_list = [10, 20, 30, 40, 50]

# 傳統方式
new_list = []
for i in my_list:
    new_list.append(i*10)

new_list
```

```python
my_list = [10, 20, 30, 40, 50]

#list的綜合表達式
new_list = [i*10 for i in my_list]
    
new_list
```

```python
# 使用numpy
import numpy as np

# 陣列運算
my_array = np.array([10, 20, 30, 40, 50])
my_array *= 10
my_array
```
