```python
import pandas as pd

# 美國球賽的小費數據集
tips = pd.read_csv('tips.csv')
tips.info()
```

```python
# 改變tips的欄位名稱
tips.columns = ['總票價', '小費', '是否吸煙', '日期', '時間', '人數']
tips
```

```python
# 重新排序欄位的順序
tips = tips[['日期', '時間', '是否吸煙', '人數', '總票價', '小費']]

# 新增一個欄位:小費比列
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
# 依照日期,建立一個群組
grouped = tips.groupby('日期')
grouped[['小費']].mean()
```

```python
# 依據日期和是否吸煙,建立一個群組
grouped1 = tips.groupby(['日期', '是否吸煙'])

# 取得小費的平均值
s1 = grouped1['小費'].mean()

# s1是一個多層索引的Series
# 變成DataFrame
df1 = s1.unstack()
df1

# s1我要unstack,日期在欄位,是否吸煙在索引
df2 = s1.unstack(level='日期')
df2
df2.T
```

```python
grouped2 = tips.groupby(['日期', '是否吸煙'])
df3 = grouped2['小費'].agg(['mean', 'sum', 'count']).unstack(level='日期')
df3.columns.names = ['統計量', '日期']

# 統計量欄位和日期欄位交換位置
df3.swaplevel(axis=1)
```

```python
grouped2 = tips.groupby(['日期', '是否吸煙'])
df5 = grouped2[['小費','總票價']].agg(['mean','sum','count'])
df5 # 此為 4維資料(四個條件組成的儲存格)
```

```python
# 改為中文
grouped2 = tips.groupby(['日期', '是否吸煙'])
df5 = grouped2[['小費','總票價']].agg([('平均','mean'),('總和','sum'),('數量','count')])
df5 
```

```python
# .agg指的是聚合
grouped3 = tips.groupby(['日期', '是否吸煙'])
grouped3[['小費']].agg([('平均','mean')])
```
